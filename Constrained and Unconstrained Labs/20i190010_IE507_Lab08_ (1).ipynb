**Q.5 Based on the previous question, estimate the number of trucks required to service all demands. What is the average number of trucks in use? What is the standard deviation in the daily number of trucks in use? Find the number of total idle truck-days in the system.**

```
Total_Trucks=ab["Central(Trucks)"]+ab["West(Trucks)"]+ab["East(Trucks)"]+ab["South(Trucks)"]+ab["North(Trucks)"]
print(Total_Trucks)
```

```
print(max(Total_Trucks))
print('The number of trucks required are:',max(Total_Trucks))
```

```

print('Average number of trucks in use are: ',np.mean(Total_Trucks))
print('Standard deviation of trucks in use: ',np.std(Total_Trucks))
```

```

Max_list=[]
for i in range(0,365):
  Max_list.append(max(Total_Trucks))

ab["Maximum trucks"]=Max_list
ab["Trucks_Idle"]=ab["Maximum trucks"]-Total_Trucks

Idle_days=0
for j in ab["Trucks_Idle"]:
  if j>0:
    Idle_days=Idle_days+1
print('The number of days in which trucks are idle: ',Idle_days)
```
